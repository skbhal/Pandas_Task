### Ques 9 Avegage response time for each agent 

```python
df.head()
```

```python
def get_seconds(time_str): #print('Time in hh:mm:ss:', time_str) 
    hh, mm, ss = time_str.split(':')  #split in hh, mm, ss
    return int(hh) * 3600 + int(mm) * 60 + int(ss)
```

```python
avg_response_sec=[]
for t in df['Average Response Time']:
    avg_response_sec.append(get_seconds(t))
avg_response_sec
df['avg_response_sec'] = avg_response_sec
```

```python
df.head(5)
```

```python
df.groupby(['Agent Name','date_week'])['avg_response_sec'].mean()
```
